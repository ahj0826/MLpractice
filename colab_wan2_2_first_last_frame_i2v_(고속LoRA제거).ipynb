```python
import os
import json
import requests
import logging
import re
from urllib.parse import urlparse
from IPython import get_ipython
import subprocess

logger = logging.getLogger(__name__)

# 모델 다운로드 설정
ENABLE_MODEL_DOWNLOAD = True
ENABLE_WAN_MODELS = True
CIVITAI_TOKEN = ""         #@param {type:"string"}

# 기본 모델 URL들 (빈 리스트에서 시작)
MODEL_URLS = {
    "checkpoints": [],
    "diffusion_models": [],
    "text_encoders": [],
    "clip": [],
    "clip_vision": [],
    "vae": [],
    "upscale_models": [],
    "sams": [],
    "loras": []
}

# WAN 모델들 추가용 URL 세트
WAN_URLS = {
    #"checkpoints": [
        #"https://huggingface.co/Phr00t/WAN2.2-14B-Rapid-AllInOne/resolve/main/wan2.2-i2v-rapid-aio.safetensors"
    #],
    "diffusion_models": [
        "https://huggingface.co/QuantStack/Wan2.2-I2V-A14B-GGUF/resolve/main/HighNoise/Wan2.2-I2V-A14B-HighNoise-Q4_K_S.gguf",
        "https://huggingface.co/QuantStack/Wan2.2-I2V-A14B-GGUF/resolve/main/LowNoise/Wan2.2-I2V-A14B-LowNoise-Q4_K_S.gguf"
    ],
    "clip_vision": [
        "https://huggingface.co/Comfy-Org/Wan_2.1_ComfyUI_repackaged/resolve/main/split_files/clip_vision/clip_vision_h.safetensors"
    ],
    "vae": [
        "https://huggingface.co/Comfy-Org/Wan_2.1_ComfyUI_repackaged/resolve/main/split_files/vae/wan_2.1_vae.safetensors"
    ],
    "loras": [
        "https://huggingface.co/Kijai/WanVideo_comfy/resolve/main/FastWan/FastWan_T2V_14B_480p_lora_rank_128_bf16.safetensors",
        "https://huggingface.co/Kijai/WanVideo_comfy/resolve/main/Lightx2v/lightx2v_T2V_14B_cfg_step_distill_v2_lora_rank32_bf16.safetensors"
    ],
    "upscale_models": [
        "https://huggingface.co/Kim2091/2x-AnimeSharpV4/resolve/main/2x-AnimeSharpV4_Fast_RCAN_PU.safetensors"
    ],
    "text_encoders": [
        "https://huggingface.co/Comfy-Org/Wan_2.1_ComfyUI_repackaged/resolve/main/split_files/text_encoders/umt5_xxl_fp8_e4m3fn_scaled.safetensors"
    ]
}


def setup_model_urls():
    """설정에 따라 MODEL_URLS에 URL들을 추가"""
    logger.info("🔧 모델 URL 설정 중...")
    # WAN 모델들 추가
    if ENABLE_WAN_MODELS:
        logger.info("  🎬 WAN 모델 URL 추가 중...")
        for category, urls in WAN_URLS.items():
            MODEL_URLS[category].extend(urls)
            if urls:  # 비어있지 않은 카테고리만 로그
                logger.info(f"    └─ {category}: {len(urls)}개 URL 추가")

    # 총 추가된 URL 개수 표시
    total_urls = sum(len(urls) for urls in MODEL_URLS.values())
    logger.info(f"  ✅ 총 {total_urls}개 모델 URL 설정 완료")

def get_civitai_url(url, token):
    """Civitai URL에 토큰 추가"""
    return f"{url}?token={token}"

def extract_filename_from_cd(header_value):
    """Content-Disposition 헤더에서 filename 파싱"""
    if not header_value:
        return None
    fname = re.findall('filename="([^"]+)"', header_value)
    if fname:
        return fname[0]
    return None

def get_filename(url, token=None):
    """다운로드할 정확한 파일명을 추출"""
    if "civitai.com" in url:
        api_url = url if not token else f"{url}?token={token}"
        try:
            response = requests.get(api_url, stream=True, allow_redirects=True)
            filename = extract_filename_from_cd(response.headers.get('Content-Disposition'))
            response.close()
            if filename:
                return filename
        except Exception as e:
            logger.error(f"파일명 요청 실패: {e}")
        # fallback: URL에서 숫자 추출
        return os.path.basename(urlparse(url).path)
    else:
        return os.path.basename(urlparse(url).path)

def download_models():
    """모델 다운로드 (Google Colab 환경에 최적화, 파일명 자동 인식)"""
    # 전체 모델 다운로드가 비활성화된 경우

    # URL 설정 먼저 수행
    setup_model_urls()

    # 다운로드할 모델이 없으면 종료
    total_urls = sum(len(urls) for urls in MODEL_URLS.values())
    if total_urls == 0:
        logger.info("📦 다운로드할 모델이 없습니다 (모든 모델 옵션이 비활성화됨)")
        return

    model_paths = {
        "checkpoints": "/content/ComfyUI/models/checkpoints",
        "diffusion_models": "/content/ComfyUI/models/unet",
        "text_encoders": "/content/ComfyUI/models/text_encoders",
        "clip": "/content/ComfyUI/models/clip",
        "clip_vision": "/content/ComfyUI/models/clip_vision",
        "vae": "/content/ComfyUI/models/vae",
        "upscale_models": "/content/ComfyUI/models/upscale_models",
        "sams": "/content/ComfyUI/models/sams",
        "loras": "/content/ComfyUI/models/loras"
    }

    logger.info("📥 모델 다운로드 시작...")

    for model_type, urls in MODEL_URLS.items():
        if not urls:  # 빈 리스트는 건너뛰기
            continue

        save_path = model_paths[model_type]
        os.makedirs(save_path, exist_ok=True)
        logger.info(f"  📂 {model_type} 모델 다운로드 중... ({len(urls)}개)")

        for i, url in enumerate(urls, 1):
            try:
                filename = get_filename(url, CIVITAI_TOKEN if "civitai.com" in url else None)
                if "civitai.com" in url:
                    download_url = get_civitai_url(url, CIVITAI_TOKEN)
                else:
                    download_url = url

                # 파일이 이미 존재하는지 확인
                full_path = os.path.join(save_path, filename)
                if os.path.exists(full_path):
                    logger.info(f"    [{i}/{len(urls)}] ⏭️ {filename} (이미 존재함)")
                    continue

                # aria2c 명령어
                cmd = f'aria2c -x 8 -s 8 --allow-overwrite=false --auto-file-renaming=false "{download_url}" -d "{save_path}" -o "{filename}"'
                logger.info(f"    [{i}/{len(urls)}] ⬇️ {filename} 다운로드 중...")
                logger.debug(f"Executing command: {' '.join(cmd)}") # 디버깅 시 명령어 확인

                # ⬇️ 2. shell=False (기본값)로 실행하고, 결과의 returncode를 확인
                result = subprocess.run(cmd, capture_output=True, text=True, shell=True, check=False)

                if result.returncode == 0:
                    logger.info(f"    [{i}/{len(urls)}] ✅ {filename} 다운로드 완료")
                else:
                    # ⬇️ 3. 실패 시 aria2c의 오류 메시지(stderr)를 함께 로깅하여 원인 파악 용이
                    logger.error(f"    [{i}/{len(urls)}] ❌ {filename} 다운로드 실패")
                    logger.error(f"    aria2c 오류: {result.stderr.strip()}")

            except FileNotFoundError:
                logger.error(f"    ❌ 'aria2c' 명령을 찾을 수 없습니다. 설치되어 있는지 확인하세요.")
            except Exception as e:
                logger.error(f"    [{i}/{len(urls)}] ❌ 모델 다운로드 중 예상치 못한 오류 발생: {e}")


def create_config_files():
    """설정 파일 생성"""
    logger.info("⚙️ 설정 파일 생성 중...")

    # rgthree 설정
    rgthree_config = {
        "features": {
            "group_header_fast_toggle": {"enabled": False},
            "import_individual_nodes": {"enabled": False},
            "menu_auto_nest": {"subdirs": False},
            "progress_bar": {"enabled": False}
        }
    }

    rgthree_dir = "/content/ComfyUI/custom_nodes/rgthree-comfy"
    os.makedirs(rgthree_dir, exist_ok=True)

    with open(os.path.join(rgthree_dir, "rgthree_config.json"), "w") as f:
        json.dump(rgthree_config, f, indent=2)

    logger.info("  ✅ rgthree 설정 파일 생성 완료")

    # 자동완성 파일 다운로드
    autocomplete_dir = "/content/ComfyUI/custom_nodes/ComfyUI-Custom-Scripts/user"
    os.makedirs(autocomplete_dir, exist_ok=True)

    try:
        response = requests.get("https://gist.githubusercontent.com/pythongosssss/1d3efa6050356a08cea975183088159a/raw/a18fb2f94f9156cf4476b0c24a09544d6c0baec6/danbooru-tags.txt")
        if response.status_code == 200:
            with open(os.path.join(autocomplete_dir, "autocomplete.txt"), "w", encoding="utf-8") as f:
                f.write(response.text)
            logger.info("  ✅ 자동완성 파일 다운로드 완료")
        else:
            logger.error(f"  ❌ 자동완성 파일 다운로드 실패: HTTP {response.status_code}")
    except Exception as e:
        logger.error(f"  ❌ 자동완성 파일 다운로드 실패: {e}")

def print_config_summary():
    """현재 설정 요약 출력"""
    logger.info("=" * 50)
    logger.info("🔧 현재 설정:")
    logger.info(f"  🎛️ 전체 모델 다운로드: {'✅ 활성화' if ENABLE_MODEL_DOWNLOAD else '❌ 비활성화'}")

    if ENABLE_MODEL_DOWNLOAD:
        logger.info(f"    └─ 🎬 WAN 모델: {'✅ 활성화' if ENABLE_WAN_MODELS else '❌ 비활성화'}")

        # 예상 다운로드 수 표시
        expected_count = 0
        if ENABLE_WAN_MODELS:
            expected_count += sum(len(urls) for urls in WAN_URLS.values())

        logger.info(f"  📊 다운로드 예정 모델: {expected_count}개")
    else:
        logger.info("  ⚠️ 모델 다운로드가 비활성화되어 있어 하위 옵션들이 무시됩니다")

    logger.info(f"  🔑 Civitai 토큰: {'설정됨' if CIVITAI_TOKEN and CIVITAI_TOKEN != 'test' else '기본값 사용'}")
    logger.info("=" * 50)

# 메인 실행부
if __name__ == "__main__":
    try:
        print_config_summary()
        download_models()
        create_config_files()
        logger.info("🎉 모든 작업이 완료되었습니다!")
    except Exception as e:
        logger.error(f"❌ 처리 중 오류 발생: {e}")
        raise
```

```python
#@title 필요 파일 생성
%%writefile /content/run_generator.py
# 이 파일은 셀 3에서 서브프로세스로 실행됩니다.

import sys
import os
import time
import glob
import gc
import torch
import subprocess
import random
import argparse
from typing import Sequence, Mapping, Any, Union

# --- 0. 기본 설정 및 인수 파싱 --
def parse_args():
    parser = argparse.ArgumentParser(description="ComfyUI Video Generation Script")
    parser.add_argument("--positive_prompt", type=str, required=True)
    parser.add_argument("--negative_prompt", type=str, required=True)
    parser.add_argument("--width", type=int, required=True)
    parser.add_argument("--height", type=int, required=True)
    parser.add_argument("--length", type=int, required=True)
    parser.add_argument("--upscale_ratio", type=float, required=True)
    parser.add_argument("--custom_lora_1_name", type=str, default="None")
    parser.add_argument("--custom_lora_1_strength_model", type=float, default=1.0)
    parser.add_argument("--custom_lora_1_strength_clip", type=float, default=1.0)
    parser.add_argument("--custom_lora_2_name", type=str, default="None")
    parser.add_argument("--custom_lora_2_strength_model", type=float, default=1.0)
    parser.add_argument("--custom_lora_2_strength_clip", type=float, default=1.0)
    parser.add_argument("--use_last_frame", action="store_true")
    return parser.parse_args()

def clear_memory():
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.cuda.ipc_collect()
    gc.collect()

# --- 1. ComfyUI 관련 함수 정의 ---
COMFYUI_BASE_PATH = '/content/ComfyUI'

def get_value_at_index(obj: Union[Sequence, Mapping], index: int) -> Any:
    try:
        return obj[index]
    except (KeyError, TypeError):
        if isinstance(obj, dict) and "result" in obj:
            return obj["result"][index]
        raise

def add_comfyui_directory_to_sys_path() -> None:
    if os.path.isdir(COMFYUI_BASE_PATH) and COMFYUI_BASE_PATH not in sys.path:
        sys.path.append(COMFYUI_BASE_PATH)
        print(f"'{COMFYUI_BASE_PATH}' added to sys.path")

def import_custom_nodes() -> None:
    try:
        import nest_asyncio
        nest_asyncio.apply()
    except ImportError:
        subprocess.run([sys.executable, "-m", "pip", "install", "-q", "nest_asyncio"])
        import nest_asyncio
        nest_asyncio.apply()

    import asyncio
    import execution
    from nodes import init_extra_nodes
    import server

    loop = asyncio.get_event_loop() or asyncio.new_event_loop()
    server_instance = server.PromptServer(loop)
    execution.PromptQueue(server_instance)
    if not loop.is_running():
        loop.run_until_complete(init_extra_nodes())
    else:
        asyncio.ensure_future(init_extra_nodes())

# --- 2. 메인 실행 로직 ---
def main():
    args = parse_args()

    print("🚀 동영상 생성을 시작합니다...")
    print(f"프롬프트: {args.positive_prompt[:50]}...")
    print(f"크기: {args.width}x{args.height}, 길이: {args.length} 프레임")
    print(f"최종 업스케일 비율: {args.upscale_ratio}x")
    if args.custom_lora_1_name != "None":
        print(f"커스텀 LoRA 1: {args.custom_lora_1_name} (강도: {args.custom_lora_1_strength_model})")
    if args.custom_lora_2_name != "None":
        print(f"커스텀 LoRA 2: {args.custom_lora_2_name} (강도: {args.custom_lora_2_strength_model})")


    subprocess.run(f"rm -rf {COMFYUI_BASE_PATH}/output/up/*", shell=True, check=True)
    subprocess.run(f"rm -rf {COMFYUI_BASE_PATH}/output/temp/*", shell=True, check=True)

    add_comfyui_directory_to_sys_path()

    from utils.extra_config import load_extra_path_config
    extra_model_paths_file = os.path.join(COMFYUI_BASE_PATH, "extra_model_paths.yaml")
    if os.path.exists(extra_model_paths_file):
        load_extra_path_config(extra_model_paths_file)

    print("Initializing ComfyUI custom nodes...")
    import_custom_nodes()
    from nodes import NODE_CLASS_MAPPINGS
    print("Custom nodes initialized successfully.")

    with torch.inference_mode():
        unetloadergguf = NODE_CLASS_MAPPINGS["UnetLoaderGGUF"]()
        cliploader = NODE_CLASS_MAPPINGS["CLIPLoader"]()
        loraloader = NODE_CLASS_MAPPINGS["LoraLoader"]()
        cliptextencode = NODE_CLASS_MAPPINGS["CLIPTextEncode"]()
        vaeloader = NODE_CLASS_MAPPINGS["VAELoader"]()
        loadimage = NODE_CLASS_MAPPINGS["LoadImage"]()
        upscalemodelloader = NODE_CLASS_MAPPINGS["UpscaleModelLoader"]()
        mxslider = NODE_CLASS_MAPPINGS["mxSlider"]()
        vhs_loadimagespath = NODE_CLASS_MAPPINGS["VHS_LoadImagesPath"]()
        modelsamplingsd3 = NODE_CLASS_MAPPINGS["ModelSamplingSD3"]()
        wanimagetovideo = NODE_CLASS_MAPPINGS["WanImageToVideo"]()
        ksampleradvanced = NODE_CLASS_MAPPINGS["KSamplerAdvanced"]()
        vaedecode = NODE_CLASS_MAPPINGS["VAEDecode"]()
        rife_vfi = NODE_CLASS_MAPPINGS["RIFE VFI"]()
        vhs_videocombine = NODE_CLASS_MAPPINGS["VHS_VideoCombine"]()
        easy_mathfloat = NODE_CLASS_MAPPINGS["easy mathFloat"]()
        imageupscalewithmodel = NODE_CLASS_MAPPINGS["ImageUpscaleWithModel"]()
        imagescaleby = NODE_CLASS_MAPPINGS["ImageScaleBy"]()
        saveimage = NODE_CLASS_MAPPINGS["SaveImage"]()


        print("Starting Step 1: Initial Sampling")
        cliploader_38 = cliploader.load_clip(clip_name="umt5_xxl_fp8_e4m3fn_scaled.safetensors", type="wan", device="default")
        unetloadergguf_84 = unetloadergguf.load_unet(unet_name="Wan2.2-I2V-A14B-HighNoise-Q4_K_S.gguf")

        # 1번 샘플러 체인

        model = get_value_at_index(unetloadergguf_84, 0)
        clip = get_value_at_index(cliploader_38, 0)
        if args.custom_lora_1_name and args.custom_lora_1_name != "None":
            print(f"1번 샘플러에 커스텀 LoRA 적용: {args.custom_lora_1_name}")
            model, clip = loraloader.load_lora(lora_name=args.custom_lora_1_name, strength_model=args.custom_lora_1_strength_model, strength_clip=args.custom_lora_1_strength_clip, model=model, clip=clip)

        cliptextencode_6 = cliptextencode.encode(text=args.positive_prompt, clip=clip) # 원본 clip 직접 연결
        cliptextencode_7 = cliptextencode.encode(text=args.negative_prompt, clip=clip)
        vaeloader_39 = vaeloader.load_vae(vae_name="wan_2.1_vae.safetensors")
        loadimage_62 = loadimage.load_image(image="example.png")

        # --- last frame 입력여부에 따른 분기 로직 ---
        if args.use_last_frame and os.path.exists(os.path.join(COMFYUI_BASE_PATH, "input", "end_example.png")):
            loadimage_end = loadimage.load_image(image="end_example.png")
            wan_node = NODE_CLASS_MAPPINGS["WanFirstLastFrameToVideo"]()
            wan_func = getattr(wan_node, "FUNCTION", "encode") if not hasattr(wan_node, "EXECUTE_NORMALIZED") else "EXECUTE_NORMALIZED"
            wanimagetovideo_63 = getattr(wan_node, wan_func)(width=args.width, height=args.height, length=args.length, batch_size=1, positive=get_value_at_index(cliptextencode_6, 0), negative=get_value_at_index(cliptextencode_7, 0), vae=get_value_at_index(vaeloader_39, 0), start_image=get_value_at_index(loadimage_62, 0), end_image=get_value_at_index(loadimage_end, 0))
        else:
            wan_node = NODE_CLASS_MAPPINGS["WanImageToVideo"]()
            wan_func = getattr(wan_node, "FUNCTION", "encode") if not hasattr(wan_node, "EXECUTE_NORMALIZED") else "EXECUTE_NORMALIZED"
            wanimagetovideo_63 = getattr(wan_node, wan_func)(width=args.width, height=args.height, length=args.length, batch_size=1, positive=get_value_at_index(cliptextencode_6, 0), negative=get_value_at_index(cliptextencode_7, 0), vae=get_value_at_index(vaeloader_39, 0), start_image=get_value_at_index(loadimage_62, 0))
        # -------------------------------------

        modelsamplingsd3_54 = modelsamplingsd3.patch(shift=8.0, model=model) # lora 대신 model 변수 직접 연결
        current_seed = int(time.time() * 1000)
        random.seed(current_seed)

        # steps=25, end_at_step=15로 변경
        ksampleradvanced_74 = ksampleradvanced.sample(add_noise="enable", noise_seed=random.randint(1, 2**64), steps=25, cfg=1, sampler_name="lcm", scheduler="simple", start_at_step=0, end_at_step=15, return_with_leftover_noise="enable", model=get_value_at_index(modelsamplingsd3_54, 0), positive=get_value_at_index(wanimagetovideo_63, 0), negative=get_value_at_index(wanimagetovideo_63, 1), latent_image=get_value_at_index(wanimagetovideo_63, 2))

        del unetloadergguf_84, modelsamplingsd3_54, loadimage_62, model, clip
        clear_memory()
        print("Step 1 finished and memory cleared.")

        print("Starting Step 2: Refinement Sampling")
        unetloadergguf_85 = unetloadergguf.load_unet(unet_name="Wan2.2-I2V-A14B-LowNoise-Q4_K_S.gguf")

        # 2번 샘플러 LoRA 체인

        model = get_value_at_index(unetloadergguf_85, 0)
        clip = get_value_at_index(cliploader_38, 0)

        if args.custom_lora_2_name and args.custom_lora_2_name != "None":
            print(f"2번 샘플러에 커스텀 LoRA 적용: {args.custom_lora_2_name}")
            model, clip = loraloader.load_lora(lora_name=args.custom_lora_2_name, strength_model=args.custom_lora_2_strength_model, strength_clip=args.custom_lora_2_strength_clip, model=model, clip=clip)

        # loraloader_86 참조를 지우고 model을 직접 patch에 전달
        modelsamplingsd3_55 = modelsamplingsd3.patch(shift=8, model=model)

        current_seed_2 = int(time.time() * 1000)
        random.seed(current_seed_2)

        # steps=25 및 start_at_step=15 로 상향 조정
        ksampleradvanced_75 = ksampleradvanced.sample(add_noise="disable", noise_seed=random.randint(1, 2**64), steps=25, cfg=1, sampler_name="lcm", scheduler="simple", start_at_step=15, end_at_step=10000, return_with_leftover_noise="disable", model=get_value_at_index(modelsamplingsd3_55, 0), positive=get_value_at_index(wanimagetovideo_63, 0), negative=get_value_at_index(wanimagetovideo_63, 1), latent_image=get_value_at_index(ksampleradvanced_74, 0))





        del unetloadergguf_85, modelsamplingsd3_55, cliploader_38, cliptextencode_6, cliptextencode_7, ksampleradvanced_74, wanimagetovideo_63, model, clip
        clear_memory()
        print("Step 2 finished and memory cleared.")

        print("Starting Step 3: VAE Decode and Save")
        vaedecode_8 = vaedecode.decode(samples=get_value_at_index(ksampleradvanced_75, 0), vae=get_value_at_index(vaeloader_39, 0))
        saveimage.save_images(filename_prefix="temp/example", images=get_value_at_index(vaedecode_8, 0))

        # ✨ [최적화] 사용이 끝난 VAE 관련 변수 즉시 삭제
        del ksampleradvanced_75, vaeloader_39, vaedecode_8
        clear_memory()
        print("Step 3 finished and memory cleared.")

        if args.upscale_ratio > 1:
            print("Starting Steps 4 & 5: Upscaling")
            upscalemodelloader_88 = upscalemodelloader.load_model(model_name="2x-AnimeSharpV4_Fast_RCAN_PU.safetensors")

            # Upscale Part 1
            vhs_loadimagespath_96 = vhs_loadimagespath.load_images(directory=f"{COMFYUI_BASE_PATH}/output/temp", image_load_cap=40)
            if get_value_at_index(vhs_loadimagespath_96, 0) is not None and get_value_at_index(vhs_loadimagespath_96, 0).shape[0] > 0:
                imageupscalewithmodel_92 = imageupscalewithmodel.upscale(upscale_model=get_value_at_index(upscalemodelloader_88, 0), image=get_value_at_index(vhs_loadimagespath_96, 0))
                mxslider_91 = mxslider.main(Xi=args.upscale_ratio, Xf=args.upscale_ratio, isfloatX=1)
                easy_mathfloat_90 = easy_mathfloat.float_math_operation(a=get_value_at_index(mxslider_91, 0), b=2, operation="divide")
                imagescaleby_93 = imagescaleby.upscale(upscale_method="nearest-exact", scale_by=get_value_at_index(easy_mathfloat_90, 0), image=get_value_at_index(imageupscalewithmodel_92, 0))
                saveimage.save_images(filename_prefix="up/example", images=get_value_at_index(imagescaleby_93, 0))
                del vhs_loadimagespath_96, imageupscalewithmodel_92, imagescaleby_93, easy_mathfloat_90, mxslider_91
                clear_memory()

            # Upscale Part 2
            vhs_loadimagespath_98 = vhs_loadimagespath.load_images(directory=f"{COMFYUI_BASE_PATH}/output/temp", skip_first_images=40)
            if get_value_at_index(vhs_loadimagespath_98, 0) is not None and get_value_at_index(vhs_loadimagespath_98, 0).shape[0] > 0:
                imageupscalewithmodel_100 = imageupscalewithmodel.upscale(upscale_model=get_value_at_index(upscalemodelloader_88, 0), image=get_value_at_index(vhs_loadimagespath_98, 0))
                mxslider_102 = mxslider.main(Xi=args.upscale_ratio, Xf=args.upscale_ratio, isfloatX=1)
                easy_mathfloat_103 = easy_mathfloat.float_math_operation(a=get_value_at_index(mxslider_102, 0), b=2, operation="divide")
                imagescaleby_101 = imagescaleby.upscale(upscale_method="nearest-exact", scale_by=get_value_at_index(easy_mathfloat_103, 0), image=get_value_at_index(imageupscalewithmodel_100, 0))
                saveimage.save_images(filename_prefix="up/example", images=get_value_at_index(imagescaleby_101, 0))
                del vhs_loadimagespath_98, imageupscalewithmodel_100, easy_mathfloat_103, imagescaleby_101, mxslider_102

            del upscalemodelloader_88
            clear_memory()
            print("Upscaling finished and memory cleared.")
            rife_input_dir = f"{COMFYUI_BASE_PATH}/output/up"
        else:
            print("Skipping Upscaling (Upscale ratio <= 1)")
            rife_input_dir = f"{COMFYUI_BASE_PATH}/output/temp"

        print("Starting Step 6: RIFE and Combine")
        vhs_loadimagespath_97 = vhs_loadimagespath.load_images(directory=rife_input_dir)
        rife_vfi_94 = rife_vfi.vfi(ckpt_name="rife47.pth", multiplier=2, frames=get_value_at_index(vhs_loadimagespath_97, 0))
        vhs_videocombine.combine_video(frame_rate=32, loop_count=0, filename_prefix="AnimateDiff", format="video/h264-mp4", pix_fmt="yuv420p", crf=19, save_metadata=True, trim_to_audio=False, pingpong=False, save_output=True, images=get_value_at_index(rife_vfi_94, 0))

        # ✨ [최적화] 마지막 단계에서 사용한 변수 삭제
        del vhs_loadimagespath_97, rife_vfi_94
        clear_memory()
        print("✅ All steps finished.")

        output_dir = os.path.join(COMFYUI_BASE_PATH, "output")
        video_files = glob.glob(os.path.join(output_dir, '**', '*.mp4'), recursive=True)

        if not video_files:
            raise FileNotFoundError("생성된 동영상 파일을 찾을 수 없습니다!")

        latest_video = max(video_files, key=os.path.getctime)
        print(f"LATEST_VIDEO_PATH:{latest_video}")

if __name__ == "__main__":
    main()
```

```python
#@title 실행
#!/usr/bin/env python3
"""
ComfyUI Video Generation Gradio WebUI
기존 Jupyter 노트북 코드를 Gradio 웹 인터페이스로 변환
"""

import gradio as gr
import subprocess
import sys
import os
import time
import glob
import gc
import torch
import random
import shutil
from pathlib import Path
from typing import Tuple, Optional
import tempfile

# ComfyUI 기본 경로 설정
COMFYUI_BASE_PATH = '/content/ComfyUI'
INPUT_DIR = os.path.join(COMFYUI_BASE_PATH, 'input')
OUTPUT_DIR = os.path.join(COMFYUI_BASE_PATH, 'output')
TEMP_OUTPUT_DIR = os.path.join(OUTPUT_DIR, 'temp')
UPSCALED_OUTPUT_DIR = os.path.join(OUTPUT_DIR, 'up')

# 디렉토리 생성
os.makedirs(INPUT_DIR, exist_ok=True)
os.makedirs(OUTPUT_DIR, exist_ok=True)
os.makedirs(TEMP_OUTPUT_DIR, exist_ok=True)
os.makedirs(UPSCALED_OUTPUT_DIR, exist_ok=True)


def clear_memory():
    """메모리 정리 함수"""
    if torch.cuda.is_available():
        torch.cuda.empty_cache()
        torch.cuda.ipc_collect()
    gc.collect()

def save_uploaded_image(image_file, end_image_file) -> str:
    """업로드된 이미지를 ComfyUI input 디렉토리에 저장"""
    if image_file is None:
        raise ValueError("이미지 파일이 선택되지 않았습니다.")

    # 이미지를 example.png로 저장
    save_path = os.path.join(INPUT_DIR, 'example.png')
    shutil.copy2(image_file, save_path)
    if end_image_file is not None:
        shutil.copy2(end_image_file, os.path.join(INPUT_DIR, 'end_example.png'))

    return save_path

# ✨ [수정됨] 함수 시그니처에 커스텀 LoRA 파라미터 추가
def run_video_generation(
    positive_prompt: str,
    negative_prompt: str,
    image_file,
    end_image_file,
    width: int,
    height: int,
    length: int,
    upscale_ratio: float,
    custom_lora_1_name: str,
    custom_lora_1_strength_model: float,
    custom_lora_1_strength_clip: float,
    custom_lora_2_name: str,
    custom_lora_2_strength_model: float,
    custom_lora_2_strength_clip: float,
    progress=gr.Progress()
) -> Tuple[Optional[str], str]:
    """
    비디오 생성 메인 함수

    Returns:
        Tuple[Optional[str], str]: (생성된 비디오 경로, 로그 메시지)
    """

    try:
        # 진행률 업데이트
        progress(0.1, "이미지 업로드 처리 중...")

        # 이미지 저장
        if image_file is None:
            return None, "❌ 오류: 이미지를 업로드해주세요."

        save_uploaded_image(image_file, end_image_file)

        # 출력 디렉토리 정리
        progress(0.2, "출력 디렉토리 정리 중...")
        subprocess.run(f"rm -rf {OUTPUT_DIR}/up/*", shell=True)
        subprocess.run(f"rm -rf {OUTPUT_DIR}/temp/*", shell=True)

        # 비디오 생성 스크립트 실행
        progress(0.3, "비디오 생성 시작...")

        start_time = time.time()

        script_path = "/content/run_generator.py"
        if not os.path.exists(script_path):
             return None, f"❌ 오류: {script_path} 파일을 찾을 수 없습니다. 이전 셀을 실행하여 파일을 생성해주세요."


        # ✨ [수정됨] 실행 명령어에 커스텀 LoRA 인자 추가
        command = [
            sys.executable,
            "-u",
            script_path,
            "--positive_prompt", positive_prompt,
            "--negative_prompt", negative_prompt,
            "--width", str(width),
            "--height", str(height),
            "--length", str(length),
            "--upscale_ratio", str(upscale_ratio),
            "--custom_lora_1_name", custom_lora_1_name,
            "--custom_lora_1_strength_model", str(custom_lora_1_strength_model),
            "--custom_lora_1_strength_clip", str(custom_lora_1_strength_clip),
            "--custom_lora_2_name", custom_lora_2_name,
            "--custom_lora_2_strength_model", str(custom_lora_2_strength_model),
            "--custom_lora_2_strength_clip", str(custom_lora_2_strength_clip),
        ]

        if end_image_file is not None:
             command.append("--use_last_frame")

        env = os.environ.copy()
        env['PYTHONUNBUFFERED'] = '1'
        env['TERM'] = 'xterm'

        process = subprocess.Popen(
            command,
            stdout=subprocess.PIPE,
            stderr=subprocess.STDOUT,
            text=True,
            encoding='utf-8',
            env=env,
            bufsize=1
        )

        # 실시간 로그 수집 및 진행률 업데이트
        full_log = []
        for line in iter(process.stdout.readline, ''):
            line = line.rstrip()
            if line:
                full_log.append(line)

                # 진행률 업데이트 (단계별)
                if "Starting Step 1" in line:
                    progress(0.4, "1단계: 초기 샘플링...")
                elif "Starting Step 2" in line:
                    progress(0.5, "2단계: 세부 조정 샘플링...")
                elif "Starting Step 3" in line:
                    progress(0.6, "3단계: VAE 디코딩 및 저장...")
                elif "Starting Steps 4 & 5" in line:
                    progress(0.7, "4-5단계: 업스케일링...")
                elif "Starting Step 6" in line:
                    progress(0.8, "6단계: RIFE 및 비디오 결합...")
                elif "All steps finished" in line:
                    progress(0.9, "마무리 중...")

        process.wait()
        return_code = process.returncode

        end_time = time.time()
        total_time = end_time - start_time
        minutes = int(total_time // 60)
        seconds = int(total_time % 60)

        progress(1.0, "완료!")

        # 결과 처리
        if return_code == 0:
            latest_video_path = None
            for line in full_log:
                if line.startswith("LATEST_VIDEO_PATH:"):
                    latest_video_path = line.replace("LATEST_VIDEO_PATH:", "").strip()
                    break

            if latest_video_path and os.path.exists(latest_video_path):
                log_message = f"""✅ 비디오 생성 완료!
⏱️ 총 소요 시간: {minutes}분 {seconds}초
📁 파일 위치: {latest_video_path}
📊 파일 크기: {os.path.getsize(latest_video_path) / (1024*1024):.2f} MB

🔧 설정 정보:
- 해상도: {width}x{height}
- 프레임 길이: {length}
- 업스케일 비율: {upscale_ratio}x
- 커스텀 LoRA 1: {custom_lora_1_name} (강도: {custom_lora_1_strength_model})
- 커스텀 LoRA 2: {custom_lora_2_name} (강도: {custom_lora_2_strength_model})
- 포지티브 프롬프트: {positive_prompt[:100]}...
- 네거티브 프롬프트: {negative_prompt[:100]}...
"""
                return latest_video_path, log_message
            else:
                error_log = "\n".join(full_log[-20:])
                return None, f"❌ 비디오 파일을 찾을 수 없습니다.\n\n마지막 로그:\n{error_log}"
        else:
            error_log = "\n".join(full_log[-20:])
            return None, f"❌ 스크ript 실행 중 오류 발생 (Return Code: {return_code})\n\n오류 로그:\n{error_log}"

    except Exception as e:
        return None, f"❌ 예상치 못한 오류 발생: {str(e)}"


def clear_output_directories():
    """출력 디렉토리 정리"""
    try:
        subprocess.run(f"rm -rf {OUTPUT_DIR}/*", shell=True)
        subprocess.run(f"rm -rf {INPUT_DIR}/*", shell=True)
        return "✅ 출력 디렉토리가 정리되었습니다."
    except Exception as e:
        return f"❌ 디렉토리 정리 중 오류: {str(e)}"

# Gradio 인터페이스 생성
def create_gradio_interface():
    """Gradio WebUI 생성"""
    # ✨ [추가됨] LoRA 파일 목록 동적 로드
    lora_dir = os.path.join(COMFYUI_BASE_PATH, 'models', 'loras')
    available_loras = ["None"] # 기본값
    if os.path.exists(lora_dir):
        try:
            lora_files = [f for f in os.listdir(lora_dir) if os.path.isfile(os.path.join(lora_dir, f)) and f.endswith(('.safetensors', '.ckpt', '.pt'))]
            available_loras.extend(lora_files)
        except Exception as e:
            print(f"LoRA 디렉토리를 읽는 중 오류 발생: {e}")

    css = """
    .gradio-container { max-width: 1200px !important; margin: auto; }
    .main-header { text-align: center; margin-bottom: 20px; padding: 20px; background: linear-gradient(45deg, #667eea 0%, #764ba2 100%); color: white; border-radius: 10px; }
    """

    with gr.Blocks(css=css, title="ComfyUI Video Generator") as interface:
        gr.HTML("<div class='main-header'><h1> WAN 2.2 14b Q4 i2v </h1><p>26.04.25 최종 업데이트</p></div>")

        with gr.Row():
            with gr.Column(scale=1):
                gr.HTML("<h3>📝 프롬프트 설정</h3>")
                positive_prompt = gr.Textbox(label="긍정 프롬프트", value="a stunning woman dancing fluidly with expressive, rhythmic motion, masterpiece, best quality, high resolution", lines=3)
                negative_prompt = gr.Textbox(label="부정 프롬프트", value="bad quality, worst quality, blurry, noisy, static, watermark, text, signature", lines=2)

                gr.HTML("<h3>🖼️ 이미지 업로드</h3>")
                image_input = gr.File(label="시작 이미지", file_types=["image"], type="filepath")
                end_image_input = gr.File(label="마지막 이미지 (선택)", file_types=["image"], type="filepath")

                gr.HTML("<h3>⚙️ 비디오 설정</h3>")
                with gr.Row():
                    width = gr.Number(label="너비 (Width)", value=480, minimum=256, maximum=1024, step=16)
                    height = gr.Number(label="높이 (Height)", value=832, minimum=256, maximum=1024, step=16)
                with gr.Row():
                    length = gr.Number(label="길이 (Frames)", value=81, minimum=16, maximum=120, step=1)
                    upscale_ratio = gr.Slider(label="업스케일 비율", minimum=1, maximum=4, step=0.1, value=2)

                # ✨ [추가됨] 커스텀 LoRA 설정 UI
                gr.HTML("<h3>🎨 커스텀 LoRA 설정 (선택 사항)</h3>")
                with gr.Group():
                    gr.Markdown("1번 샘플러 (초기 생성) LoRA")
                    custom_lora_1_name = gr.Dropdown(label="LoRA 선택", choices=available_loras, value="None")
                    with gr.Row():
                        custom_lora_1_strength_model = gr.Slider(label="Model 강도", minimum=0.0, maximum=2.0, step=0.05, value=1.0)
                        custom_lora_1_strength_clip = gr.Slider(label="Clip 강도", minimum=0.0, maximum=2.0, step=0.05, value=1.0)

                with gr.Group():
                    gr.Markdown("2번 샘플러 (품질 개선) LoRA")
                    custom_lora_2_name = gr.Dropdown(label="LoRA 선택", choices=available_loras, value="None")
                    with gr.Row():
                        custom_lora_2_strength_model = gr.Slider(label="Model 강도", minimum=0.0, maximum=2.0, step=0.05, value=1.0)
                        custom_lora_2_strength_clip = gr.Slider(label="Clip 강도", minimum=0.0, maximum=2.0, step=0.05, value=1.0)


                with gr.Row():
                    generate_btn = gr.Button("🚀 비디오 생성", variant="primary", size="lg")
                    clear_btn = gr.Button("🗑️ 입출력 정리", variant="secondary")

            with gr.Column(scale=1):
                gr.HTML("<h3>📹 생성 결과</h3>")
                video_output = gr.Video(label="생성된 비디오", height=400)
                log_output = gr.Textbox(label="처리 로그", lines=15, max_lines=20, interactive=False)

        # ✨ [수정됨] 이벤트 핸들러에 새로운 입력 추가
        all_inputs = [
            positive_prompt, negative_prompt, image_input, end_image_input, width, height, length, upscale_ratio,
            custom_lora_1_name, custom_lora_1_strength_model, custom_lora_1_strength_clip,
            custom_lora_2_name, custom_lora_2_strength_model, custom_lora_2_strength_clip
        ]
        generate_btn.click(
            fn=run_video_generation,
            inputs=all_inputs,
            outputs=[video_output, log_output],
            show_progress=True
        )

        clear_btn.click(fn=clear_output_directories, inputs=[], outputs=[log_output])

        with gr.Accordion("📖 사용법 안내", open=False):
             gr.Markdown("...") # 내용은 동일하므로 생략

    return interface

def main():
    print("🚀 ComfyUI Video Generator WebUI 시작...")
    if not os.path.exists(COMFYUI_BASE_PATH):
        print(f"❌ ComfyUI 경로를 찾을 수 없습니다: {COMFYUI_BASE_PATH}")
        return
    interface = create_gradio_interface()
    interface.launch(server_name="0.0.0.0", server_port=7860, share=True, debug=True, show_error=False, inbrowser=False, allowed_paths=[OUTPUT_DIR])

if __name__ == "__main__":
    print("🔍 시스템 환경 확인...")
    if torch.cuda.is_available():
        print(f"✅ CUDA 사용 가능: {torch.cuda.get_device_name(0)}")
        print(f"   VRAM: {torch.cuda.get_device_properties(0).total_memory / 1024**3:.1f}GB")
    else:
        print("⚠️  CUDA를 사용할 수 없습니다. CPU 모드로 실행됩니다.")
    for directory in [COMFYUI_BASE_PATH, INPUT_DIR, OUTPUT_DIR]:
        if os.path.exists(directory):
            print(f"✅ 디렉토리 확인: {directory}")
        else:
            print(f"📁 디렉토리 생성: {directory}")
            os.makedirs(directory, exist_ok=True)
    main()
```
